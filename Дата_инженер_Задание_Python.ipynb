**Задание 0**

Определить вероятность выпадения цифры 8 на десятом броске 9-гранного кубика. Ответ укажите в комментариях в файле ipynb, где буду выполняться следующие задания.

Решу задачу двумя способами:

1) Математическое решение

```python
def prob_eight(): # обозначаем функцию, которая будет бросать кубики
    proba = 1 / 9  # Вероятность выпадения 8 на одном броске девятигранного кубика
    return proba

print("Вероятность выпадения цифры 8 (не важно, какой бросок):", prob_eight())
```

2) Решение при помощи рандомных бросков (ЭКСПЕРИМЕНТ)

```python
import random

def proba_eight(tr_all, tr_10): # обозначаем функцию, которая будет бросать кубики
    count_eight = 0 # счетчик выпадений 8 на 10-м броске
    for roll in range(tr_all): # бросаем кубик 100000 раз
        val = random.randint(1, 9) # получаем случайное значение
        if val == 8 and (roll + 1) % 10 == 0: # проверяем, что значение кубика 8 и бросок у нас 10-й по счету
            count_eight += 1 # записываем тот случай, когда выпало число 8

    prob = count_eight / tr_10 # находим вероятность через отношение
    return prob

trials = 100000
ten_trials = 10000 # десятый бросок выполнен 10000 раз (каждый десятый бросок из 100000 попыток)
print(f"Вероятность выпадения цифры 8 на десятом броске: {proba_eight(trials, ten_trials)}")
```

Нужно посчитать количество заказов по каждой номенклатуре

```python
# Вывожу страницу ИП Иванова, чтобы иметь перед глазами пример таблицы
df_ivanov
```

```python
# Обозначаю тип данных столбца заказов по умолчанию. Использую тип Int64 для большей точности (если вдруг что)
df['Заказы, шт.'] = df['Заказы, шт.'].astype('Int64')

# Счетчик индекса на главной таблице
count_num = 0


# Беру одно значение номенклатуры по порядку
for nom in df['Номенклатура']:

    # Считываю имя ИП
    name_ind = df.loc[count_num, 'ИП']

    # Работаю с условием, чтобы точно понимать, какому ИП принадлежит номенклатура
    if name_ind == 'Иванов':

        # Помещаю все индексы одной номенклатуры в список
        index = df_ivanov.index[df_ivanov['Unnamed: 5'] == nom].tolist()

        # Считаю заказы
        count_ord = 0

        # Перебираю индексы
        for ind in index:

            # Фиксирую количество товара
            count_ord += df_ivanov.loc[ind, 'Заказано']

        # Записываю финальное количество товара в главную таблицу
        df.loc[count_num, 'Заказы, шт.'] = count_ord

    # Аналогично действую с другими ИП

    elif name_ind == 'Петров':
        index = df_petrov.index[df_petrov['Unnamed: 5'] == nom].tolist()
        count_ord = 0
        for ind in index:
            count_ord += df_petrov.loc[ind, 'Заказано']
        df.loc[count_num, 'Заказы, шт.'] = count_ord
    elif name_ind == 'Сидоров':
        index = df_sidorov.index[df_sidorov['Unnamed: 5'] == nom].tolist()
        count_ord = 0
        for ind in index:
            count_ord += df_sidorov.loc[ind, 'Заказано']
        df.loc[count_num, 'Заказы, шт.'] = count_ord

    # Если ИП не найдено, то записываю 0
    else:
        df.loc[count_num, 'Заказы, шт.'] = 0

    # Не забываю считать индексы
    count_num += 1
```

1.3

Нужно посчитать выручку по каждой номенклатуре.

```python
# Счетчик индекса на главной странице
count_num = 0


# Беру одно значение номенклатуры по порядку
for nom in df['Номенклатура']:

    # Считываю имя ИП
    name_ind = df.loc[count_num, 'ИП']

    # Работаю с условием, чтобы точно понимать, какому ИП принадлежит номенклатура
    if name_ind == 'Иванов':

        # Помещаю все индексы одной номенклатуры в список
        index = df_ivanov.index[df_ivanov['Unnamed: 5'] == nom].tolist()

        # Считаю выручку
        count_ord = 0

        # Перебираю индексы
        for ind in index:

            # Записываю значение по каждому индексу
            count_ord += df_ivanov.loc[ind, 'Unnamed: 11']

        # При подсчете выручки использую модуль decimal для точности при работе с типом float
        arg_shot = Decimal(str(count_ord + count_ord * 0.17))

        # Записываю полученные значение в главную таблицу, предвварительно округлив их до двух знаков после запятой
        df.loc[count_num, 'Выручка, руб.'] = float(arg_shot.quantize(Decimal('1.00')))

    # Аналогичные действия выполняю с другими ИП
    elif name_ind == 'Петров':
        index = df_petrov.index[df_petrov['Unnamed: 5'] == nom].tolist()
        count_ord = 0
        for ind in index:
            count_ord += df_petrov.loc[ind, 'Unnamed: 11']
        arg_shot = Decimal(str(count_ord + count_ord * 0.17))
        df.loc[count_num, 'Выручка, руб.'] = float(arg_shot.quantize(Decimal('1.00')))
    elif name_ind == 'Сидоров':
        index = df_sidorov.index[df_sidorov['Unnamed: 5'] == nom].tolist()
        count_ord = 0
        for ind in index:
            count_ord += df_sidorov.loc[ind, 'Unnamed: 11']
        arg_shot = Decimal(str(count_ord + count_ord * 0.17))
        df.loc[count_num, 'Выручка, руб.'] = float(arg_shot.quantize(Decimal('1.00')))
    else:
        # Если ИП не найдено, то записываю 0
        df.loc[count_num, 'Выручка, руб.'] = 0

    # Считаю индексы
    count_num += 1
```

1.4

Необходимо посчитать прибыль по каждой номенклатуре.

```python
# Загружаю таблицу "Себестоимость и Цена до скидки"
df_cost = pd.read_excel("/content/drive/MyDrive/Дата-инженер_Юдин_Владислав/Данные для задания Python.xlsx", sheet_name='Себестоимость и Цена до скидки')

# Загружаю таблицу "Спаравочник"
df_guide = pd.read_excel("/content/drive/MyDrive/Дата-инженер_Юдин_Владислав/Данные для задания Python.xlsx", sheet_name='Справочник')

# Проверяю, загрущилась ли таблица
df_cost
```

```python
# Счетчик индексов на главной странице
count_num = 0


# Беру значение номенклатуры по порядку
for nom in df['Номенклатура']:

    # Считываю имя ИП
    name_ind = df.loc[count_num, 'ИП']

    # Работаю с условием, чтобы точно понимать, какому ИП принадлежит номенклатура
    if name_ind == 'Иванов':

        # Добавляю условие: если заказов нет, то и прибыль равна нулю
        if df.loc[count_num, 'Заказы, шт.'] == 0:
              df.loc[count_num, 'Прибыль, руб.'] = 0
        else:

            # В таблице Справочник нахожу индекс текущей номенклатуры
            ind_guide = df_guide[df_guide['Номенклатура'] == nom].index[0]

            # Записываю категорию номенклатуры из Справочника
            guide = df_guide.loc[ind_guide, 'Категория']

            # Использую исключения, следуя приммечанию из задания
            # Примечание: если себестоимость не указана для какой-то из категорий,
            #то задаем ее равной минимальной себестоимости из таблицы с себестоимостями.
            try:
                # Условие, если категорию возможно найти
                ind_cost = df_cost[df_cost['Категория'] == guide].index[0]
                cost = df_cost.loc[ind_cost, 'Фиксированные затраты, руб./шт.']
            except:
                # Задаем минимальное значение, если категория отсутствует
                cost = min(df_cost['Фиксированные затраты, руб./шт.'])

            # Получаю значение выручки по конкретной номенклатуре, чтобы далее использовать ее в формуле
            revenue = df.loc[count_num, 'Выручка, руб.']

            # Высчитываю прибыль
            gain = Decimal(str(revenue - (cost + revenue * 0.01 + revenue * 0.17)))

            # Записываю полученный результат в главную таблицу
            df.loc[count_num, 'Прибыль, руб.'] = float(gain.quantize(Decimal('1.00')))

    # Аналогично для других ИП
    elif name_ind == 'Петров':

        if df.loc[count_num, 'Заказы, шт.'] == 0:
              df.loc[count_num, 'Прибыль, руб.'] = 0
        else:
            ind_guide = df_guide[df_guide['Номенклатура'] == nom].index[0]
            guide = df_guide.loc[ind_guide, 'Категория']
            try:
                ind_cost = df_cost[df_cost['Категория'] == guide].index[0]
                cost = df_cost.loc[ind_cost, 'Фиксированные затраты, руб./шт.']
            except:
                cost = min(df_cost['Фиксированные затраты, руб./шт.'])


            revenue = df.loc[count_num, 'Выручка, руб.']
            gain = Decimal(str(revenue - (cost + revenue * 0.03 + revenue * 0.17)))

            df.loc[count_num, 'Прибыль, руб.'] = float(gain.quantize(Decimal('1.00')))

    elif name_ind == 'Сидоров':

        if df.loc[count_num, 'Заказы, шт.'] == 0:
            df.loc[count_num, 'Прибыль, руб.'] = 0
        else:
            ind_guide = df_guide[df_guide['Номенклатура'] == nom].index[0]
            guide = df_guide.loc[ind_guide, 'Категория']
            try:
                ind_cost = df_cost[df_cost['Категория'] == guide].index[0]
                cost = df_cost.loc[ind_cost, 'Фиксированные затраты, руб./шт.']
            except:
                cost = min(df_cost['Фиксированные затраты, руб./шт.'])


            revenue = df.loc[count_num, 'Выручка, руб.']
            gain = Decimal(str(revenue - (cost + revenue * 0.05 + revenue * 0.17)))

            df.loc[count_num, 'Прибыль, руб.'] = float(gain.quantize(Decimal('1.00')))

    else:
        # 0 для номенклатуры без ИП
        df.loc[count_num, 'Прибыль, руб.'] = 0

    # Счетчик индексов
    count_num += 1
```

Нужно посчитать рентабельность продаж по каждой номенклатуре.

```python
# Счетчик индексов на главной странице
count_num = 0

# Рентабельности присваиваю целочисленное значение
df['Рентабельность, %'] = df['Рентабельность, %'].astype('Int64')


# Беру значение номенклатуры по порядку
for nom in df['Номенклатура']:

    # Задаю условие, чтобы не работать с номенклатурой, которая не принадлежит ни одному ИП
    if df.loc[count_num, 'Выручка, руб.'] != 0:

        # Высчитываю рентабельность по формуле
        profit = Decimal(str(df.loc[count_num, 'Прибыль, руб.']/df.loc[count_num, 'Выручка, руб.']))

        # Перевожу результат в проценты
        profit_pr = int(profit.quantize(Decimal('1.00')) * 100)

        # Записываю рентабельность в главную таблицу
        df.loc[count_num, 'Рентабельность, %'] = profit_pr

        # Считаю индекс
        count_num += 1
    else:

        # Записываю нулевое значение
        df.loc[count_num, 'Рентабельность, %'] = 0

        # Считаю индекс
        count_num += 1
```

```python
# Счетчик индекса
count_num = 0

# Значение номенклатуры по порядку
for nom in df_2['Номенклатура']:
    # Фиксированные затраты
    fix = df_2.loc[count_num, 'Фиксированные затрат, руб./шт.']
    # Налог
    tax = df_2.loc[count_num, 'Налог, %']
    # Комиссия
    commission = df_2.loc[count_num, 'Комиссия,% ']
    # Цена после скидки
    price_disc = df_2.loc[count_num, 'Цена после скидки']

    # Вычисляем затраты на единицу
    unit_price = Decimal(str(fix)) + Decimal(str(tax + commission)) / Decimal(100) * Decimal(str(price_disc))

    # Вычисляем итоговую рентабельность
    profit = (Decimal(str(price_disc)) - unit_price) / Decimal(str(price_disc))

    # Заносим рентабельность в таблицу
    df_2.loc[count_num, 'Рентабельность, %'] = profit.quantize(Decimal('1.00')) * Decimal(100)

    # Счетчик индекса
    count_num += 1
```
